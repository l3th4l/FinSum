```python
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt 
import seaborn as sns

import datetime

import os
```

```python
import statsmodels as stm
import statsmodels.api as sm
from statsmodels.api import tsa 
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
cur_codes = [x[:-4] for x in os.listdir('./Dataset')]
```

```python
cur_codes
```

```python
dat = pd.read_csv('./Dataset/USD.csv', parse_dates = True)
#preprocess dates
dat['Date'] = [d.split(' ')[0] for d in dat['Date']]
dat['Date'] = [datetime.datetime(int(d.split('/')[2]), int(d.split('/')[0]), int(d.split('/')[1])) for d in dat['Date']]
dat['Year'] = [x.year for x in dat.Date]
dat['Month'] = [x.month for x in dat.Date]
dat = dat[["Date", "Year", "Month", "Close"]]
dat
```

```python
#convert to monthly data
dat_m = dat.groupby(["Year", "Month"], as_index = False).mean()
dat_m["Date"] = [datetime.datetime(x[0], x[1], 1) for x in zip(dat_m.Year, dat_m.Month)]
dat_m
```

```python
plt.figure(figsize = [24, 8])
sns.lineplot(data = dat_m.reset_index(), x = "Date", y = 'Close')
```

```python
stl.adfuller(dat_m["Close"])
```

```python
res = sm.tsa.seasonal_decompose(dat_m["Close"], period = 12, model =  "additive")
plot = res.plot()
plot.set(figwidth = 20, figheight = 20)
```

```python
residual = res.resid.dropna()
```

```python
plt.figure(figsize = [20, 6])
plt.plot(range(len(residual)), residual - np.mean(residual))
plt.scatter(range(len(residual)), residual - np.mean(residual))
```

```python
stl.adfuller(residual)
```

```python
plt.figure(figsize = [20, 6])
acf = tsa.acf(residual)
plt.scatter(range(len(acf)), acf, 100)
plt.vlines(range(len(acf)),np.zeros(len(acf)),ymax= acf)
plt.title("acf")
```

```python
plt.figure(figsize = [20, 6])
pacf = tsa.pacf(residual)
plt.scatter(range(len(pacf)), pacf, 100)
plt.vlines(range(len(pacf)),np.zeros(len(pacf)),ymax= pacf)
plt.title("pacf")
```

```python
#deseasonalize 
plt.figure(figsize = [20, 6])
plt.plot(range(len(residual)), residual - np.mean(residual))
plt.scatter(range(len(residual)), residual - np.mean(residual))
```
